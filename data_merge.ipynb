1. Completing train datas with other necessary features

```python
pivot1 = pd.merge(based_train, edu, on='user_id', how='left')
pivot2 = pd.merge(pivot1, exp, on='user_id', how='left')
pivot3 = pd.merge(pivot2, lang, on='user_id', how='left')
pivot4 = pd.merge(pivot3, skills, on='user_id', how='left')
pivot4.head()
```
