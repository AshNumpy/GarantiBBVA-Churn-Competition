```python
# libs 
import pandas as pd 
import numpy as np 
```

```python
train = pd.read_csv('../ML-DL Datasets/train.csv')
test = pd.read_csv('../ML-DL Datasets/test.csv')
train.head()
```

```python
spec_cols = ['school_count', 'experience_month', 'exp_comp_count', 'language_count', 'skill_count', 'moved_after_2019', 'industry', 'location']
train[spec_cols].head()
```

```python
# scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
train[['school_count', 'experience_month', 'exp_comp_count', 'language_count', 'skill_count']] = sc.fit_transform(train[['school_count', 'experience_month', 'exp_comp_count', 'language_count', 'skill_count']])
train[['school_count', 'experience_month', 'exp_comp_count', 'language_count', 'skill_count']].head()
```

```python
# label encoder
from sklearn.preprocessing import LabelEncoder
industry_le = LabelEncoder()
location_le = LabelEncoder()
```

```python
train['industry'] = industry_le.fit_transform(train['industry'])
train['location'] = location_le.fit_transform(train['location'])
train[['industry', 'location']].head()
```

```python
train[spec_cols].head()
```

```python
X_train = train.drop(['user_id', 'moved_after_2019'], axis=1)
y_train = train[['moved_after_2019']]
```

```python
test[['school_count', 'experience_month', 'exp_comp_count', 'language_count', 'skill_count']] = sc.transform(test[['school_count', 'experience_month', 'exp_comp_count', 'language_count', 'skill_count']])
test[['school_count', 'experience_month', 'exp_comp_count', 'language_count', 'skill_count']].head()
```

```python
test['industry'] = industry_le.transform(test['industry'])
test['location'] = location_le.transform(test['location'])
test[['industry', 'location']].head()
```

```python
user_id = test[['user_id']]
X_test = test.drop(['user_id'], axis=1)
```

```python
X_train.to_csv('../Boosts/Datasets/X_train.csv', index=False)
y_train.to_csv('../Boosts/Datasets/y_train.csv', index=False)
X_test.to_csv('../Boosts/Datasets/X_test.csv', index=False)
user_id.to_csv('../Boosts/Datasets/test_user_id.csv', index=False)
```
