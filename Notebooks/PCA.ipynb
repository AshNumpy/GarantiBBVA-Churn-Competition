```python
# libs
import pandas as pd
import numpy as np 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

<hr>

### Education

```python
education = pd.read_csv('../Cleaned Datasets/education.csv')
education.head()
```

```python
edu_scaler = StandardScaler()
scaled_data = edu_scaler.fit_transform(education.loc[:, education.columns != 'user_id'])
scaled_data[:1]
```

```python
edu_pca = PCA()

# PCA uygulama
edu_pca.fit(scaled_data)

# PCA sonuçlarını görselleştirme
plt.plot(np.cumsum(edu_pca.explained_variance_ratio_))
plt.xlabel('Component Number')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```
